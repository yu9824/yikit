### モジュールのインポート

```python
from yikit.feature_selection import WrapperSelector
from sklearn.datasets import load_boston
import pandas as pd
```

### サンプルデータの取得

```python
boston = load_boston()
X = pd.DataFrame(boston['data'], columns = boston['feature_names'])
y = pd.Series(boston['target'], name = 'PRICE')
X.head()
```

### 削減してみる

```python
selector = WrapperSelector(random_state = 334)
selector.fit(X, y)
X_selected = X.loc[:, selector.get_support()]
X_selected.head()
```
